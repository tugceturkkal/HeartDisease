```python
##Importing the relevant libraries
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
sns.set()
```

```python
#Loading the raw data

raw_data = pd.read_csv("heart.csv")

raw_data.head()
```

```python
#Exploring the descriptive statistics of the variables

raw_data.describe(include = "all")
```

```python
#I will create the regression without 'sex','cp','fbs','restecg','exang','slope', 'ca', 'thal' and 'target'

data = raw_data.drop(['sex','cp','fbs','restecg','exang','slope', 'ca', 'thal','target'], axis = 1)
data.describe(include = "all")
```

```python
#Dealing with missing values
data.isnull().sum()
```

```python
#Exploring the PDFs (Probability Distribution Function)
sns.histplot(data["age"],kde=True)
```

```python
sns.histplot(data["trestbps"],kde=True)
```

```python
#Let's declare a variable that will be equal to the 99th percentile of the 'trestbps' variable
q = data["trestbps"].quantile(0.99)

data_1 = data[data["trestbps"] < q]

data_1.describe(include = "all")
```

```python
sns.histplot(data_1["trestbps"],kde=True)
```

```python
sns.histplot(data["chol"], kde=True)
```

```python
q = data["chol"].quantile(0.99)

data_2 = data_1[data_1["chol"] < q]

data_2.describe(include = "all")
```

```python
sns.histplot(data_2["chol"], kde=True)
```

```python
sns.histplot(data["thalach"], kde=True)
```

```python
q = data["thalach"].quantile(0.01)

data_3 = data_2[data_2["thalach"] > q]

data_3.describe(include = "all")
```

```python
sns.histplot(data_3["thalach"], kde=True)
```

```python
sns.histplot(data["oldpeak"], kde=True)
```

```python
q = data["oldpeak"].quantile(0.99)

data_4 = data_3[data_3["oldpeak"] < q]

data_4.describe(include = "all")
```

```python
sns.histplot(data_4["oldpeak"], kde=True)
```

```python
data_cleaned = data_4.reset_index(drop = True)
data_cleaned
```
