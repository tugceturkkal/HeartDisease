```python
##Importing the relevant libraries
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
sns.set()
```

```python
#Loading the raw data

raw_data = pd.read_csv("heart.csv")

raw_data.head()
```

```python
#Exploring the descriptive statistics of the variables

raw_data.describe(include = "all")
```

```python
#I will create the regression without 'sex','cp','fbs','restecg','exang','slope', 'ca', 'thal' and 'target'

data = raw_data.drop(['sex','cp','fbs','restecg','exang','slope', 'ca', 'thal','target'], axis = 1)
data.describe(include = "all")
```

```python
data_cleaned = data_4.reset_index(drop = True)
data_cleaned
```

```python
data_cleaned.describe(include='all')
```

```python
#Checked the correlation
data_cleaned.corr()
```

```python
#visualize the correlation
sns.pairplot(data_cleaned, kind  ="reg")
```

```python
#Linear regression model
#Declare the inputs and the targets

# The target(s) (dependent variable) is 'age'
targets = data_cleaned['age']

# The inputs are everything BUT the dependent variable, so we can simply drop it
inputs = data_cleaned.drop(['age'],axis=1)
```

```python
#Scale the data 
#import the scaling module
from sklearn.preprocessing import StandardScaler

#create a scaler object
scaler = StandardScaler()

# Fit the inputs (calculate the mean and standard deviation feature-wise)
scaler.fit(inputs)
```

```python
# Scale the features and store them in a new variable (the actual scaling procedure)
inputs_scaled = scaler.transform(inputs)
```

```python
#Train Test Split
#import the module for the split
from sklearn.model_selection import train_test_split

# Split the variables with an 80-20 split and some random state
# To have the same split as mine, use random_state = 136
x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets , test_size = 0.2, random_state = 136)
```

```python
#Create the regression
#create a linear regression object 
reg = LinearRegression()

# Fit the regression with the scaled TRAIN inputs and targets
reg.fit(x_train, y_train)
```

```python
# Let's check the outputs of the regression 
#I'll store them in y_hat as this is the 'theoretical' name of the predictions
y_hat = reg.predict(x_train)
```

```python
# The simplest way to compare the targets (y_train) and the predictions (y_hat) is to plot them on a scatter plot
plt.scatter(y_train, y_hat)
# Let's also name the axes
plt.xlabel('Targets (y_train)',size=18)
plt.ylabel('Predictions (y_hat)',size=18)
plt.show()
```

```python
# Another useful check of our model is a residual plot
# We can plot the PDF of the residuals and check for anomalies
sns.histplot(y_train - y_hat, kde = True)

# Include a title
plt.title("Residuals PDF", size=18)

```

```python
# Find the R-squared of the model
reg.score(x_train,y_train)
```
